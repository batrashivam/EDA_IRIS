```python
# A variable named iris has been formed and the csv file has been loaded in it using the pandas library 
iris = pd.read_csv("iris.csv")
```

```python
# Here we are counting the total number of entries on the basis of the number of species 
iris["species"].value_counts()
```

```python
# Here we are plotting a bar graph having species on the X axis and the number of observations on the Y axis
plt.title("Count of Species")
plt.xlabel("Types of Species")
plt.ylabel("Count")
sns.countplot(iris["species"], )
```

```python
# We'll use sklearn library to make a scatterplot of the Iris features.
sns.scatterplot(iris["sepal_length"],	iris["sepal_width"], color="red")
```

```python
# We'll use sklearn library to make a scatterplot of the Iris features.
iris.plot(kind="scatter", x="petal_length", y="petal_width", color="green")
```

```python
# We can also use the seaborn library to make a similar plot that containg the scatter plot as well as a histogram
# A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x = "sepal_length", y = "sepal_width", data=iris, size=7, color="red")
```

```python
# We can also use the seaborn library to make a similar plot that containg the scatter plot as well as a histogram
# A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x = "petal_length", y = "petal_width", data=iris, size=7, color="green")
```

```python
#Here we are coloring the scatter plot on the basis of species for sepal length
sns.FacetGrid(iris, hue="species", size=6)\
  .map(plt.scatter, "sepal_length", "sepal_width")\
  .add_legend()
```

```python
#Here we are coloring the scatter plot on the basis of species for petal length
sns.FacetGrid(iris, hue="species", size=6)\
  .map(plt.scatter, "petal_length", "petal_width")\
  .add_legend()
```

```python
# We can look at an individual feature in Seaborn through a boxplot
#Here we have created  a box plot to see the spread of the dataset
sns.boxplot(x="species", y="petal_length",data=iris)
```

```python
# We can look at an individual feature in Seaborn through a boxplot
#Here we have created  a box plot to see the spread of the dataset
sns.boxplot(x="species", y="sepal_length",data=iris)
```

```python
# We'll use jitter=True so that all the points don't fall in single vertical lines for petal length
ax = sns.boxplot(x="species", y="petal_length",data=iris)
ax = sns.stripplot(x="species", y="petal_length",data=iris, jitter=True, edgecolor="Grey")
```

```python
# We'll use jitter=True so that all the points don't fall in single vertical lines for sepal length
ax = sns.boxplot(x="species", y="sepal_length",data=iris)
ax = sns.stripplot(x="species", y="sepal_length",data=iris, jitter=True, edgecolor="Grey")
```

```python
#Here we have created a vilon plot
# A violin plot combines the benefits of the previous two plots and simplifies them
# Denser regions of the data are fatter, and sparser thiner in a violin plot
#On X axis we have the species and on y axis we have the petal length
sns.violinplot(x=iris["species"], y=iris["petal_length"], size=6)
```

```python
sns.kdeplot(data=iris)
```

```python
# A final seaborn plot useful for looking at univariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(data=iris, hue="species", size=6)\
  .map(sns.kdeplot, "petal_length")\
  .add_legend()
```

```python
# A final seaborn plot useful for looking at univariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(data=iris, hue="species", size=6)\
  .map(sns.kdeplot, "sepal_length")\
  .add_legend()
```

```python
#PairPlot
sns.pairplot(data=iris, size=5, hue="species")
```

```python
# The diagonal elements in a pairplot show the histogram by default
# We can update these elements to show other things, such as a kde
sns.pairplot(data=iris, hue="species", size=3, diag_kind="kde")
```

```python
# Now that we've covered seaborn, let's go back to some of the ones we can make with Pandas
# We can quickly make a boxplot with Pandas on each feature split out by species
iris.boxplot(by="species", figsize=(12,12))
```
